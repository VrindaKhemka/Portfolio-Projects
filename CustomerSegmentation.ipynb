```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

from scipy.cluster.hierarchy import dendrogram, linkage, cophenet
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("C:/Users/vrind/OneDrive/Documents/NU/Python/CC GENERAL.csv", index_col='CUST_ID')
df.head(20)
```

```python
mask = np.isfinite(df).all(axis=1)
data = df[mask]
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(data)
normalized_data
```

```python
distances = pdist(normalized_data)
threshold = np.mean(distances)
single_linkage = linkage(normalized_data, method='single')
complete_linkage = linkage(normalized_data, method='complete')
ward_linkage = linkage(normalized_data, method='ward')
```

```python
coph_corr_single = cophenet(single_linkage, distances)
coph_corr_complete = cophenet(complete_linkage, distances)
coph_corr_ward = cophenet(ward_linkage, distances)
```

```python
print(f'Cophenetic correlation for Single Linkage: {coph_corr_single}')
print(f'Cophenetic correlation for Complete Linkage: {coph_corr_complete}')
print(f'Cophenetic correlation for Ward Linkage: {coph_corr_ward}')
```

```python
num_clusters_single = len(np.where(single_linkage[:, 2] > threshold)[0]) + 1
num_clusters_complete = len(np.where(complete_linkage[:, 2] > threshold)[0]) + 1
num_clusters_ward = len(np.where(ward_linkage[:, 2] > threshold)[0]) + 1

print(f'Number of clusters for Single Linkage: {num_clusters_single}')
print(f'Number of clusters for Complete Linkage: {num_clusters_complete}')
print(f'Number of clusters for Ward Linkage: {num_clusters_ward}')
```
